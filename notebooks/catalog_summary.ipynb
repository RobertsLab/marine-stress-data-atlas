# Marine Stress Data Atlas — catalog summary

Quick coverage overview of `data/dataset_catalog.csv`: how many datasets, which
species/stressors/omics types are represented, and where the gaps are.

Requires `pandas` and `matplotlib`:

```bash
pip install pandas matplotlib
```

> Example rows (`status == 'example'`) are excluded from the counts below.

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

# Resolve the catalog path whether run from repo root or notebooks/
here = Path.cwd()
candidates = [here / 'data' / 'dataset_catalog.csv',
              here.parent / 'data' / 'dataset_catalog.csv']
catalog_path = next((p for p in candidates if p.exists()), candidates[0])

df = pd.read_csv(catalog_path)
real = df[df['status'].str.lower() != 'example'].copy()
print(f'Loaded {len(df)} rows ({len(real)} real, {len(df) - len(real)} example) from {catalog_path}')
real.head()
````

## Milestone progress

Target: **100 datasets across at least 20 species.**

````python
n_datasets = len(real)
n_species = real['species'].nunique()
print(f'Datasets: {n_datasets} / 100  ({n_datasets}% of the way there)')
print(f'Species:  {n_species} / 20')
print(f'Reviewed: {(real["status"].str.lower() == "reviewed").sum()} rows')
````

## Coverage by category

````python
for col in ['stressor', 'omics_type', 'life_stage', 'data_repository']:
    print(f'\n== {col} ==')
    print(real[col].value_counts(dropna=False).to_string())
````

````python
# Top species
top_species = real['species'].value_counts().head(15)
print(top_species.to_string())
````

## Where are the gaps? (stressor × omics)

Empty cells are candidate tasks for students to fill.

````python
if len(real):
    pivot = pd.crosstab(real['stressor'], real['omics_type'])
    display(pivot)
else:
    print('No real datasets yet — add some rows to see the coverage matrix.')
````

## Charts

````python
if len(real):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    real['stressor'].value_counts().plot.bar(ax=axes[0], title='Datasets by stressor')
    real['omics_type'].value_counts().plot.bar(ax=axes[1], title='Datasets by omics type')
    for ax in axes:
        ax.set_xlabel('')
    plt.tight_layout()
    plt.show()
else:
    print('No real datasets yet.')
````
